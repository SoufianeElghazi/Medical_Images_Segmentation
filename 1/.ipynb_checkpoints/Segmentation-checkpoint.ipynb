## Segmentation basée sur l’intensité (Thresholding)


#### 1. Seuillage global :

```python
import matplotlib.pyplot as plt
import numpy as np

# Sélection des seuils pour chaque canal
seuil = 127

# Binarisation pour chaque canal
image_binarisee_R = (image_originale[:,:,0] > seuil).astype(np.uint8)
image_binarisee_G = (image_originale[:,:,1] > seuil).astype(np.uint8)
image_binarisee_B = (image_originale[:,:,2] > seuil).astype(np.uint8)

# Affichage des trois figures
plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(image_binarisee_R, cmap='gray')
plt.title("Canal Rouge")

plt.subplot(132)
plt.imshow(image_binarisee_G, cmap='gray')
plt.title("Canal Vert")

plt.subplot(133)
plt.imshow(image_binarisee_B, cmap='gray')
plt.title("Canal Bleu")

plt.tight_layout()
plt.show()
```

#### 2. Seuillage adaptatif :

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import threshold_local

# Sélection des paramètres pour la binarisation adaptative
block_size = 35
offset = 10

# Binarisation adaptative pour chaque canal
seuil_adaptatif_R = threshold_local(image_originale[:,:,0], block_size=block_size, offset=offset)
seuil_adaptatif_G = threshold_local(image_originale[:,:,1], block_size=block_size, offset=offset)
seuil_adaptatif_B = threshold_local(image_originale[:,:,2], block_size=block_size, offset=offset)

image_binarisee_R = (image_originale[:,:,0] > seuil_adaptatif_R).astype(np.uint8)
image_binarisee_G = (image_originale[:,:,1] > seuil_adaptatif_G).astype(np.uint8)
image_binarisee_B = (image_originale[:,:,2] > seuil_adaptatif_B).astype(np.uint8)

# Affichage des trois figures
plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(image_binarisee_R, cmap='gray')
plt.title("Binarisation adaptative - Canal Rouge")

plt.subplot(132)
plt.imshow(image_binarisee_G, cmap='gray')
plt.title("Binarisation adaptative - Canal Vert")

plt.subplot(133)
plt.imshow(image_binarisee_B, cmap='gray')
plt.title("Binarisation adaptative - Canal Bleu")

plt.tight_layout()
plt.show()
```

#### 3. Seuillage d’Otsu :

```python
import matplotlib.pyplot as plt
import numpy as np
from skimage.filters import threshold_otsu

# Binarisation d'Otsu pour chaque canal
seuil_otsu_R = threshold_otsu(image_originale[:,:,0])
seuil_otsu_G = threshold_otsu(image_originale[:,:,1])
seuil_otsu_B = threshold_otsu(image_originale[:,:,2])

image_binarisee_R = (image_originale[:,:,0] > seuil_otsu_R).astype(np.uint8)
image_binarisee_G = (image_originale[:,:,1] > seuil_otsu_G).astype(np.uint8)
image_binarisee_B = (image_originale[:,:,2] > seuil_otsu_B).astype(np.uint8)

# Affichage des trois figures
plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(image_binarisee_R, cmap='gray')
plt.title("Binarisation d'Otsu - Canal Rouge")

plt.subplot(132)
plt.imshow(image_binarisee_G, cmap='gray')
plt.title("Binarisation d'Otsu - Canal Vert")

plt.subplot(133)
plt.imshow(image_binarisee_B, cmap='gray')
plt.title("Binarisation d'Otsu - Canal Bleu")

plt.tight_layout()
plt.show()
```

#### 4. Seuillage multi-niveaux :

```python
import matplotlib.pyplot as plt
import numpy as np

# Seuils pour la segmentation
seuil1 = 85
seuil2 = 170

# Création des masques pour chaque canal
mask1_R = (image_originale[:,:,0] <= seuil1)
mask2_R = (image_originale[:,:,0] > seuil1) & (image_originale[:,:,0] <= seuil2)
mask3_R = (image_originale[:,:,0] > seuil2)

mask1_G = (image_originale[:,:,1] <= seuil1)
mask2_G = (image_originale[:,:,1] > seuil1) & (image_originale[:,:,1] <= seuil2)
mask3_G = (image_originale[:,:,1] > seuil2)

mask1_B = (image_originale[:,:,2] <= seuil1)
mask2_B = (image_originale[:,:,2] > seuil1) & (image_originale[:,:,2] <= seuil2)
mask3_B = (image_originale[:,:,2] > seuil2)

# Attribution des classes pour chaque canal
image_segmentee_R = mask1_R + 2 * mask2_R + 3 * mask3_R
image_segmentee_G = mask1_G + 2 * mask2_G + 3 * mask3_G
image_segmentee_B = mask1_B + 2 * mask2_B + 3 * mask3_B

# Affichage des trois figures
plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(image_segmentee_R, cmap='nipy_spectral')
plt.title("Canal Rouge")

plt.subplot(132)
plt.imshow(image_segmentee_G, cmap='nipy_spectral')
plt.title("Canal Vert")

plt.subplot(133)
plt.imshow(image_segmentee_B, cmap='nipy_spectral')
plt.title("Canal Bleu")

plt.tight_layout()
plt.show()
```

## Segmentation par croissance de région (Region Growing)


#### Implémentation de base de la croissance de région  :

```python
import numpy as np

def region_growing(image, seed):
    height, width = image.shape
    visited = np.zeros_like(image, dtype=bool)
    segmented = np.zeros_like(image, dtype=bool)
    list_pixels = [seed]
    intensity_threshold = 5  # Ajuster selon l’image

    while len(list_pixels) > 0:
        current_pixel = list_pixels.pop()
        x, y = current_pixel

        if visited[y, x]:
            continue

        visited[y, x] = True
        segmented[y, x] = True

        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < width and 0 <= ny < height:
                    difference = abs(int(image[y, x]) - int(image[ny, nx]))
                    if not visited[ny, nx] and difference < intensity_threshold:
                        list_pixels.append((nx, ny))

    return segmented
```

```python
# Séparer les canaux de couleur
red_channel = image_originale[:,:,0]
green_channel = image_originale[:,:,1]
blue_channel = image_originale[:,:,2]

# Afficher les histogrammes pour chaque canal de couleur
plt.hist(red_channel.ravel(), bins=256, range=(0, 255), color='r', alpha=0.5, label='Rouge')
plt.hist(green_channel.ravel(), bins=256, range=(0, 255), color='g', alpha=0.5, label='Vert')
plt.hist(blue_channel.ravel(), bins=256, range=(0, 255), color='b', alpha=0.5, label='Bleu')
plt.title('Histogramme de l\'image en couleur')
plt.xlabel('Valeur de pixel')
plt.ylabel('Fréquence')
plt.legend(loc='upper right')
plt.show()
```

#### Application sur une image :

```python
# Appliquer la segmentation par croissance de région à chaque canal de couleur
seed_point = (50, 50)  # Choisir en fonction de l’image
segmented_R = region_growing(image_originale[:, :, 0], seed_point)
segmented_G = region_growing(image_originale[:, :, 1], seed_point)
segmented_B = region_growing(image_originale[:, :, 2], seed_point)

# Afficher les images segmentées pour chaque canal de couleur
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(segmented_R, cmap='gray')
plt.title("Segmentation - Canal Rouge")

plt.subplot(132)
plt.imshow(segmented_G, cmap='gray')
plt.title("Segmentation - Canal Vert")

plt.subplot(133)
plt.imshow(segmented_B, cmap='gray')
plt.title("Segmentation - Canal Bleu")

plt.tight_layout()
plt.show()
```

##  Segmentation par Clustering:


#### Segmentation d’une image avec K-means :

```python
from sklearn.cluster import KMeans

# Redimensionnement de l’image en une liste de pixels
image_reshaped = image_originale.reshape((-1, 3))  # pour une image couleur
k = 3  # Nombre de clusters , ajuster selon les besoins
kmeans = KMeans(n_clusters=k)
kmeans.fit(image_reshaped)
clustered = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = clustered.reshape(image_originale.shape)

# Afficher les images segmentées pour chaque canal de couleur

plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(segmented_image[:, :, 0].astype(np.uint8), cmap='gray')
plt.title("Segmentation - Canal Rouge")

plt.subplot(132)
plt.imshow(segmented_image[:, :, 1].astype(np.uint8), cmap='gray')
plt.title("Segmentation - Canal Vert")

plt.subplot(133)
plt.imshow(segmented_image[:, :, 2].astype(np.uint8), cmap='gray')
plt.title("Segmentation - Canal Bleu")

plt.tight_layout()
plt.show()
```

##  Segmentation basée sur la couleur


#### Segmentation d’une image en utilisant le modèle de couleur HSV :


pip install opencv-python

```python
import cv2

# Fonction pour segmenter une image basée sur la couleur HSV
def segment_by_color_hsv(image, lower_bound, upper_bound):
    # Convertir l’image de BGR à HSV
    hsv_image = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)

    # Appliquer un masque basé sur les limites de couleur spécifiées
    mask = cv2.inRange(hsv_image, lower_bound, upper_bound)

    # Utiliser le masque pour extraire la région d’intérêt de l’image originale
    segmented_image = cv2.bitwise_and(image, image, mask=mask)

    return segmented_image

# Définir les limites de couleur pour la teinte verte
lower = np.array([30, 50, 50])
upper = np.array([60, 255, 255])

# Segmenter l’image originale en utilisant les limites définies
segmented_image = segment_by_color_hsv(image_originale, lower, upper)

# Afficher les composantes de l'image segmentée pour chaque canal de couleur HSV
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 3))

plt.subplot(131)
plt.imshow(segmented_image[:, :, 0], cmap='gray')
plt.title("Canal Hue")

plt.subplot(132)
plt.imshow(segmented_image[:, :, 1], cmap='gray')
plt.title("Canal Saturation")

plt.subplot(133)
plt.imshow(segmented_image[:, :, 2], cmap='gray')
plt.title("Canal Value")

plt.tight_layout()
plt.show()
```

##  Exploitation de Datasets Médicaux


```python
import cv2
import numpy as np
from skimage import io, color
from sklearn.cluster import KMeans
from skimage.filters import threshold_local
import matplotlib . pyplot as plt


# Fonctions de segmentation
#def thresholding(image):
#    #image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#    _, segmented_image = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
#    return segmented_image

def thresholding_adaptatif(image, block_size=11, constant=2):
    if image.ndim == 3 and image.shape[-1] == 3:
        # L'image est en couleurs, convertissez-la en niveaux de gris
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)       
    # Appliquer le seuillage adaptatif
    segmented_image = cv2.adaptiveThreshold(image, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, block_size, constant)  
    return segmented_image


def region_growing(image, seed=(5, 5)):
    if len(image.shape) == 3 and image.shape[2] == 3:  # Vérifier si l'image a trois canaux
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    height, width = image.shape
    visited = np.zeros_like(image, dtype=bool)
    segmented = np.zeros_like(image, dtype=bool)
    list_pixels = [seed]
    intensity_threshold = 5  # Ajuster selon l’image

    while len(list_pixels) > 0:
        current_pixel = list_pixels.pop()
        x, y = current_pixel

        if visited[y, x]:
            continue

        visited[y, x] = True
        segmented[y, x] = True

        for dx in [-1, 0, 1]:
            for dy in [-1, 0, 1]:
                nx, ny = x + dx, y + dy
                if 0 <= nx < width and 0 <= ny < height:
                    difference = abs(int(image[y, x]) - int(image[ny, nx]))
                    if not visited[ny, nx] and difference < intensity_threshold:
                        list_pixels.append((nx, ny))

    return segmented


# Chemins des dossiers contenant les images
dossiers = [
    "Images/Images Cérébrales/pictures/",
    "Images/Images de Mammographie/pictures/",
    "Images/Images Rétiniennes/pictures/",
    "Images/Tumeurs Cérébrales/pictures/"
]

# Liste des méthodes de segmentation
methodes = [
    #thresholding,
    thresholding_adaptatif,
    region_growing, 
]


```

### Segmentation basée sur l’intensité

#### ~Images de Mammographie:

```python
# Initialiser un conteneur pour stocker les résultats
resultats = []

print(f"Comparatif pour le dossier : {dossiers[1]}")
# Pour chaque méthode
for i, methode in enumerate(methodes):
    print(f"\tMéthode {i+1}:")
    jaccard_scores = []
    dice_scores = []
    for j in range(1, 13):  
        # Charger l'image
        chemin_image = f"{dossiers[1]}{j}.png"
        image = io.imread(chemin_image)
        #image = image[:,:,2] 

        # Appliquer la méthode de segmentation
        segmented_image = methode(image)

            
        # Enregistrement de l'image segmentée (dans un dossier 'resultats' par exemple)
        chemin_resultat = f"Images/Images de Mammographie/Results/{j}_meth_{i+1}.png"
        plt.imshow(segmented_image)  # Affiche l'image
        plt.axis('off')  # Désactive les axes
        plt.savefig(chemin_resultat, bbox_inches='tight', pad_inches=0, dpi=300)

        # Évaluation quantitative (coefficient de Jaccard)
        
        # Charger la vérité terrain
        ground_truth_path = "Images/Images de Mammographie/Verite_Terrain/verite_terrain.png"
        ground_truth = cv2.imread(ground_truth_path, cv2.IMREAD_GRAYSCALE)
        # la vérité terrain a la même taille que l'image segmentée
        ground_truth = cv2.resize(ground_truth, (segmented_image.shape[1], segmented_image.shape[0]))
        
        # Évaluation quantitative (coefficient de Jaccard)
        intersection = np.logical_and(ground_truth, segmented_image)
        union = np.logical_or(ground_truth, segmented_image)
        jaccard_score = np.sum(intersection) / np.sum(union)
        jaccard_scores.append(jaccard_score)
        
        # Évaluation quantitative (indice de Dice)
        intersection = np.logical_and(ground_truth, segmented_image)
        union = np.logical_or(ground_truth, segmented_image)
        dice_score = 2 * np.sum(intersection) / np.sum(union)
        dice_scores.append(dice_score)
            
    # Calculer la moyenne des scores de Jaccard et de Dice
    moyenne_jaccard = np.mean(jaccard_scores)
    moyenne_dice = np.mean(dice_scores)
    print(f"\tScore de Jaccard moyen : {moyenne_jaccard}")
    print(f"\tIndice de Dice moyen : {moyenne_dice}")
    resultats.append((dossiers[1], i+1, moyenne_jaccard, moyenne_dice))

    print("\t---")
```

##  Comparatif des Méthodes de Segmentation


#### Méthodes Évaluées


    ______________________________________________________________________________________________________________________
    Seuillage Binaire
    -----------------
    --------Forces :
    Simple et rapide et Convient aux images présentant des contrastes élevés entre les structures.
    --------Faiblesses :
    Sensible aux variations d'éclairage.
    Ne fonctionne pas bien avec des images à faible contraste.
    ______________________________________________________________________________________________________________________
    Seuillage de Otsu
    -----------------
    --------Forces :
    Détermine automatiquement un seuil adapté.
    Efficace pour les images bimodales.
    --------Faiblesses :
    Peut produire des résultats médiocres pour les images non bimodales.
    ______________________________________________________________________________________________________________________
    Seuillage Adaptatif
    -----------------
    --------Forces :
    S'adapte aux variations locales d'intensité.
    Convient aux images présentant des variations d'éclairage.
    --------Faiblesses :
    Requiert le réglage de paramètres tels que la taille du bloc et la constante.
    ______________________________________________________________________________________________________________________
    K-Means Clustering
    -----------------
    --------Forces :
    Peut identifier automatiquement les clusters de pixels.
    Utile pour les images avec une distribution de couleur complexe.
    --------Faiblesses :
    Requiert une initialisation correcte des centroïdes.
    ______________________________________________________________________________________________________________________
    Region Growing
    -----------------
    --------Forces :
    Segmentations précises en fonction de la région d'amorçage.
    Utile pour les structures avec des contours bien définis.
    --------Faiblesses :
    Sensible au choix de la graine initiale.
    ______________________________________________________________________________________________________________________

#### Pertinence Selon les Types d'Images

    ~ Seuillage Binaire : Convient aux images avec des contrastes élevés, comme les images T1.
    ______________________________________________________________________________________________________________________
    ~ Seuillage de Otsu : Efficace pour les images avec une distribution de pixel bimodale, par exemple, les images T2.
     ______________________________________________________________________________________________________________________
    ~ Seuillage Adaptatif : Recommandé pour les images présentant des variations locales d'intensité, comme les images IRM pondérées en T2.
     ______________________________________________________________________________________________________________________
    ~ K-Means Clustering : Adapté aux images avec des distributions de couleur complexes, comme les images IRM FLAIR.
     ______________________________________________________________________________________________________________________
    ~ Region Growing : Utile pour les structures avec des contours bien définis, comme les tumeurs cérébrales.
     ______________________________________________________________________________________________________________________

#### Particularités des Structures Segmentées
    ~Les tumeurs cérébrales ont été mieux segmentées par la méthode de Region Growing en raison de leurs contours définis.
    ______________________________________________________________________________________________________________________
    ~ Les ventricules et les zones de fluides ont été bien identifiés par le Seuillage Adaptatif en raison des variations locales d'intensité.
